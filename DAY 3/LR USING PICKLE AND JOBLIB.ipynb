```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score
import pickle
import joblib
df = pd.read_csv("safenet_synthetic_dataset_10000.csv")
print(df.head())
X = df[['Crowd_Score']]
y = df['Threat_Risk']
model = LinearRegression()
model.fit(X, y)
prediction = model.predict([[75]])
print("Predicted Threat Risk:", prediction[0])
y_pred = model.predict(X)
mae = mean_absolute_error(y, y_pred)
r2 = r2_score(y, y_pred)
print("Mean Absolute Error:", mae)
print("R2 Score:", r2)
with open("linear_model_pickle.pkl", "wb") as file:
    pickle.dump(model, file)
joblib.dump(model, "linear_model_joblib.joblib")
with open("linear_model_pickle.pkl", "rb") as file:
    pickle_model = pickle.load(file)
print("Pickle Model Loaded!")
joblib_model = joblib.load("linear_model_joblib.joblib")
print("Joblib Model Loaded!")
pickle_prediction = pickle_model.predict([[85]])
print("Pickle Prediction:", pickle_prediction[0])
joblib_prediction = joblib_model.predict([[85]])
print("Joblib Prediction:", joblib_prediction[0])
```
